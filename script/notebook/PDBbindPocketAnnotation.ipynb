## Dataset Pockets Lable

- ✅Pocket Protein List comes from the dataset file.

```python
import pickle
import torch
from tqdm import tqdm
import pandas as pd
import numpy as np
from torchdrug import data, utils

path = "../../../data/dta-datasets/PDBbind/"
dataset_file = "../../../data/dta-datasets/PDBbind/pdbbind_datasets.csv"
pocket_pkl = path + "gearnet_pocket_Protein.pkl"
protein_pkl = path + "gearnet_Protein.pkl"
```

```python
dataset_df = pd.read_csv(dataset_file)
dataset_df
```

```python
with utils.smart_open(protein_pkl, "rb") as fin:
    protein_list  = pickle.load(fin)
```

```python
with utils.smart_open(pocket_pkl, "rb") as fin:
    pocket_list  = pickle.load(fin)
```

```python
len(pocket_list)
```

```python
pocket_list[0]
```

```python
pocket_residues = pocket_list[0].residue_number.tolist()
pocket_residues
```

```python
len(pocket_residues)
```

```python
label_list = []
# indexes = range(5)
indexes = range(dataset_df["PDB_ID"].shape[0])
indexes = tqdm(indexes, "Constructing Pocket from PKL...: ")
for index in indexes:
    label_dict = {}
    current_protein = protein_list[index]
    current_pocket = pocket_list[index]

    current_length = len(current_protein.residue_number)
    label_dict['mask'] = np.ones(current_length, dtype='int32') 
    label_dict['target'] = np.zeros(current_length, dtype='int32')  

    protein_residues = current_protein.residue_number.tolist()
    pocket_residues = current_pocket.residue_number.tolist()
    indices = [protein_residues.index(x) for x in pocket_residues]
    for i in indices:
        label_dict['target'][i] = 1
    label_list.append(label_dict)
```
